**Q2**

```python
import matplotlib.pyplot as plt
```

```python
exponential_data = np.random.exponential(scale=1.0,size=(100000,1))
uniform_data = np.random.rand(100000)
normal_data = np.random.rand(100000)
```

plot histrogram

```python
plt.hist(exponential_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-3, 6, 0, 1.2])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

**Q3**

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.title('3D plot of Z = X^2 + Y^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

df.head()

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Calculate correlations
pearson_corr = df[features].corr(method='pearson')
spearman_corr = df[features].corr(method='spearman')

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=ax[0], cbar_kws={'label': 'Pearson Correlation'})
ax[0].set_title('Pearson Correlation Coefficients')
ax[0].set_xticklabels(features, rotation=45)
ax[0].set_yticklabels(features, rotation=45)

sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=ax[1], cbar_kws={'label': 'Spearman Correlation'})
ax[1].set_title('Spearman Correlation Coefficients')
ax[1].set_xticklabels(features, rotation=45)
ax[1].set_yticklabels(features, rotation=45)

plt.tight_layout()
plt.show()
```
